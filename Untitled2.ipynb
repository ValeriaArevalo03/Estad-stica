**1.** Se desea estimar la altura promedio de estudiantes universitarios. En una muestra de 50 estudiantes,
se obtuvo una media de 170 cm. Se sabe que la desviación estándar poblacional es de 10 cm.
Construye un intervalo de confianza del 95 % para la media poblacional.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# --- Paso 1: Definir los parámetros del problema ---
media_muestra = 170
desviacion_estandar_poblacional = 10
tamano_muestra = 50
nivel_confianza = 0.95
alpha = 1 - nivel_confianza

# --- Paso 2: Calcular el valor crítico Z ---
# El valor crítico para un intervalo de dos colas se encuentra en alpha/2
z_critico = norm.ppf(1 - alpha / 2)
print(f"Valor crítico Z: {z_critico:.4f}")

# --- Paso 3: Calcular el margen de error ---
margen_error = z_critico * (desviacion_estandar_poblacional / np.sqrt(tamano_muestra))
print(f"Margen de error: {margen_error:.4f} cm")

# --- Paso 4: Calcular el intervalo de confianza ---
limite_inferior = media_muestra - margen_error
limite_superior = media_muestra + margen_error
print(f"Intervalo de Confianza del 95%: ({limite_inferior:.2f}, {limite_superior:.2f}) cm")

# --- Paso 5: Graficar la distribución y el intervalo ---
# Crear un rango de valores para la distribución normal
x = np.linspace(media_muestra - 4 * margen_error, media_muestra + 4 * margen_error, 1000)
# Calcular la función de densidad de probabilidad (PDF)
pdf = norm.pdf(x, loc=media_muestra, scale=desviacion_estandar_poblacional / np.sqrt(tamano_muestra))

plt.figure(figsize=(10, 6))
plt.plot(x, pdf, label='Distribución Muestral de la Media')
plt.axvline(media_muestra, color='r', linestyle='--', label='Media de la Muestra (170 cm)')

# Rellenar el área del intervalo de confianza
x_intervalo = np.linspace(limite_inferior, limite_superior, 100)
pdf_intervalo = norm.pdf(x_intervalo, loc=media_muestra, scale=desviacion_estandar_poblacional / np.sqrt(tamano_muestra))
plt.fill_between(x_intervalo, pdf_intervalo, color='skyblue', alpha=0.5, label='Intervalo de Confianza del 95%')

# Marcar los límites y el valor crítico
plt.axvline(limite_inferior, color='g', linestyle='-.', label=f'Límite Inferior ({limite_inferior:.2f})')
plt.axvline(limite_superior, color='g', linestyle='-.', label=f'Límite Superior ({limite_superior:.2f})')

plt.title('Intervalo de Confianza del 95% para la Altura Promedio de Estudiantes')
plt.xlabel('Altura (cm)')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t

# --- Datos del Ejercicio 9 ---
# Tiempo promedio de la muestra
media_muestra_9 = 30
# Desviación estándar de la muestra
desviacion_estandar_muestra_9 = 4
# Tamaño de la muestra
tamano_muestra_9 = 20
# Nivel de confianza
nivel_confianza_9 = 0.95
alpha_9 = 1 - nivel_confianza_9
# Grados de libertad
grados_libertad_9 = tamano_muestra_9 - 1

# --- 1. Calcular el valor crítico t para un 95% de confianza ---
t_critico_9 = t.ppf(1 - alpha_9 / 2, df=grados_libertad_9)
print(f"Ejercicio 9 - Valor crítico t: {t_critico_9:.4f}")

# --- 2. Calcular el margen de error ---
margen_error_9 = t_critico_9 * (desviacion_estandar_muestra_9 / np.sqrt(tamano_muestra_9))
print(f"Ejercicio 9 - Margen de error: {margen_error_9:.4f} minutos")

# --- 3. Calcular el intervalo de confianza ---
limite_inferior_9 = media_muestra_9 - margen_error_9
limite_superior_9 = media_muestra_9 + margen_error_9
print(f"Ejercicio 9 - IC del 95%: ({limite_inferior_9:.2f}, {limite_superior_9:.2f}) minutos")

# --- 4. Graficar la distribución t y el intervalo ---
x_9 = np.linspace(media_muestra_9 - 4 * margen_error_9, media_muestra_9 + 4 * margen_error_9, 1000)
pdf_9 = t.pdf(x_9, df=grados_libertad_9, loc=media_muestra_9, scale=desviacion_estandar_muestra_9 / np.sqrt(tamano_muestra_9))

plt.figure(figsize=(10, 6))
plt.plot(x_9, pdf_9, label=f'Distribución t ({grados_libertad_9} gl)')
plt.axvline(media_muestra_9, color='r', linestyle='--', label='Media de la Muestra (30 min)')
plt.fill_between(x_9, pdf_9, where=(x_9 >= limite_inferior_9) & (x_9 <= limite_superior_9),
                 color='skyblue', alpha=0.5, label='Intervalo de Confianza del 95%')
plt.axvline(limite_inferior_9, color='g', linestyle='-.', label=f'Límite Inferior ({limite_inferior_9:.2f})')
plt.axvline(limite_superior_9, color='g', linestyle='-.', label=f'Límite Superior ({limite_superior_9:.2f})')
plt.title('IC del 95% para el Tiempo Promedio de Entrega de Pizza')
plt.xlabel('Tiempo (minutos)')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

**21.** Una empresa afirma que sus paquetes de cereal tienen un peso promedio de 500 g. Se sospecha que
la máquina está descalibrada y subllenando. Se toma una muestra de 40 paquetes, obteniendo una
media de 495 g. La desviación estándar poblacional es σ = 15 g. Con un nivel de de significancia
α = 0,05.

*1.* Plantee las hipótesis nula y alternativa.

*2.* Calcule el estadístico Z.

*3.* Determine la región crítica y la región de rechazo.

*4.* Concluya si se rechaza H0.

```python
# ==============================================================================
# Problema #21: Prueba de Hipótesis para la Media de una Muestra
# ==============================================================================

# ------------------------------------------------------------------------------
# 1. Importación de Bibliotecas
# ------------------------------------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns

# ------------------------------------------------------------------------------
# 2. Definición de las Hipótesis y Parámetros
# ------------------------------------------------------------------------------
# Hipótesis Nula (H₀): La media poblacional es igual o mayor a 500 g. (µ ≥ 500)
# Hipótesis Alternativa (Hₐ): La media poblacional es menor a 500 g. (µ < 500)
# Se trata de una prueba de una cola (inferior).

# Parámetros del problema
media_poblacional_ho = 500  # Valor hipotetizado en H₀
media_muestral = 495       # Media de la muestra
desviacion_estandar_poblacional = 15  # Desviación estándar de la población (σ)
tamano_muestra = 40        # n
alpha = 0.05               # Nivel de significancia

# ------------------------------------------------------------------------------
# 3. Cálculo del Estadístico Z y el p-valor
# ------------------------------------------------------------------------------
# Fórmula del Estadístico Z: Z = (x̄ - µ₀) / (σ / √n)
error_estandar = desviacion_estandar_poblacional / np.sqrt(tamano_muestra)
z_calculado = (media_muestral - media_poblacional_ho) / error_estandar

# p-valor para una prueba de una cola (inferior)
p_valor = stats.norm.cdf(z_calculado)

# ------------------------------------------------------------------------------
# 4. Determinación del Valor Crítico y la Región de Rechazo
# ------------------------------------------------------------------------------
# Para una prueba de una cola (inferior), el valor crítico Z deja un área de α
# en la cola izquierda de la distribución normal.
z_critico = stats.norm.ppf(alpha)

# ------------------------------------------------------------------------------
# 5. Visualización del Resultado
# ------------------------------------------------------------------------------
# Configuración del gráfico
sns.set(style="whitegrid")
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)

plt.figure(figsize=(12, 7))
plt.plot(x, y, color='blue', label='Distribución Normal Estándar')

# Región de rechazo (área en la cola izquierda)
plt.fill_between(x, y, where=(x < z_critico), color='red', alpha=0.5, label='Región de Rechazo (α = 0.05)')

# Líneas para el valor crítico y el estadístico calculado
plt.axvline(z_critico, color='red', linestyle='--', label=f'Valor Crítico Z = {z_critico:.4f}')
plt.axvline(z_calculado, color='green', linestyle='-', label=f'Estadístico Z Calculado = {z_calculado:.4f}')

# Etiquetas y título
plt.title('Prueba de Hipótesis para la Media (Una Cola Inferior)')
plt.xlabel('Valor de Z')
plt.ylabel('Densidad de Probabilidad')
plt.text(z_critico - 0.7, 0.05, 'Rechazar H₀', color='red', rotation=90)
plt.text(z_critico + 0.1, 0.05, 'No rechazar H₀', color='blue', rotation=90)
plt.legend()
plt.grid(True)
plt.show()

# ------------------------------------------------------------------------------
# 6. Conclusión
# ------------------------------------------------------------------------------
print("==============================================================================")
print("Resultados de la Prueba de Hipótesis")
print("==============================================================================")
print(f"Estadístico Z calculado: {z_calculado:.4f}")
print(f"Valor crítico Z: {z_critico:.4f}")
print(f"p-valor: {p_valor:.4f}")
print("------------------------------------------------------------------------------")

if p_valor < alpha:
    print(f"Decisión: Se rechaza la hipótesis nula (H₀).")
    print(f"El p-valor ({p_valor:.4f}) es menor que el nivel de significancia ({alpha}).")
    print(f"Conclusión: Existe evidencia significativa para afirmar que el peso promedio de los paquetes es menor a 500 g.")
else:
    print(f"Decisión: No se rechaza la hipótesis nula (H₀).")
    print(f"El p-valor ({p_valor:.4f}) es mayor o igual que el nivel de significancia ({alpha}).")
    print(f"Conclusión: No hay suficiente evidencia para afirmar que el peso promedio de los paquetes es menor a 500 g.")
```

```python
# ==============================================================================
# Problema: Prueba de Hipótesis para la Duración de Baterías
# ==============================================================================

# ------------------------------------------------------------------------------
# 1. Importación de Bibliotecas
# ------------------------------------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns

# ------------------------------------------------------------------------------
# 2. Definición de las Hipótesis y Parámetros
# ------------------------------------------------------------------------------
# El cliente alega que la duración es menor, lo que implica una prueba de una cola (inferior).
# Hipótesis Nula (H₀): La media poblacional es igual o mayor a 1200 horas. (µ ≥ 1200)
# Hipótesis Alternativa (Hₐ): La media poblacional es menor a 1200 horas. (µ < 1200)

# Parámetros del problema
media_poblacional_ho = 1200  # Valor hipotetizado en H₀
media_muestral = 1180       # Media de la muestra
desviacion_estandar_poblacional = 100  # Desviación estándar de la población (σ)
tamano_muestra = 50        # n
alpha = 0.01               # Nivel de significancia

# ------------------------------------------------------------------------------
# 3. Cálculo del Estadístico Z y el p-valor
# ------------------------------------------------------------------------------
# Fórmula del Estadístico Z: Z = (x̄ - µ₀) / (σ / √n)
error_estandar = desviacion_estandar_poblacional / np.sqrt(tamano_muestra)
z_calculado = (media_muestral - media_poblacional_ho) / error_estandar

# p-valor para una prueba de una cola (inferior)
p_valor = stats.norm.cdf(z_calculado)

# ------------------------------------------------------------------------------
# 4. Determinación del Valor Crítico y la Región de Rechazo
# ------------------------------------------------------------------------------
# Para una prueba de una cola (inferior), el valor crítico Z deja un área de α
# en la cola izquierda de la distribución normal.
z_critico = stats.norm.ppf(alpha)

# ------------------------------------------------------------------------------
# 5. Visualización del Resultado
# ------------------------------------------------------------------------------
# Configuración del gráfico
sns.set(style="whitegrid")
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)

plt.figure(figsize=(12, 7))
plt.plot(x, y, color='blue', label='Distribución Normal Estándar')

# Región de rechazo (área en la cola izquierda)
plt.fill_between(x, y, where=(x < z_critico), color='red', alpha=0.5, label=f'Región de Rechazo (α = {alpha})')

# Líneas para el valor crítico y el estadístico calculado
plt.axvline(z_critico, color='red', linestyle='--', label=f'Valor Crítico Z = {z_critico:.4f}')
plt.axvline(z_calculado, color='green', linestyle='-', label=f'Estadístico Z Calculado = {z_calculado:.4f}')

# Etiquetas y título
plt.title('Prueba de Hipótesis para la Duración de Baterías (Una Cola Inferior)')
plt.xlabel('Valor de Z')
plt.ylabel('Densidad de Probabilidad')
plt.text(z_critico - 0.7, 0.05, 'Rechazar H₀', color='red', rotation=90)
plt.text(z_critico + 0.1, 0.05, 'No rechazar H₀', color='blue', rotation=90)
plt.legend()
plt.grid(True)
plt.show()

# ------------------------------------------------------------------------------
# 6. Conclusión
# ------------------------------------------------------------------------------
print("==============================================================================")
print("Resultados de la Prueba de Hipótesis")
print("==============================================================================")
print(f"Estadístico Z calculado: {z_calculado:.4f}")
print(f"Valor crítico Z: {z_critico:.4f}")
print(f"p-valor: {p_valor:.4f}")
print("------------------------------------------------------------------------------")

if p_valor < alpha:
    print(f"Decisión: Se rechaza la hipótesis nula (H₀).")
    print(f"El p-valor ({p_valor:.4f}) es menor que el nivel de significancia ({alpha}).")
    print(f"Conclusión: Con un nivel de significancia del 1%, existe evidencia estadística para respaldar la afirmación del cliente.")
    print("Es decir, la duración promedio de las baterías es significativamente menor a 1200 horas.")
else:
    print(f"Decisión: No se rechaza la hipótesis nula (H₀).")
    print(f"El p-valor ({p_valor:.4f}) es mayor o igual que el nivel de significancia ({alpha}).")
    print(f"Conclusión: No hay suficiente evidencia estadística para respaldar la afirmación del cliente.")
    print("No se puede concluir que la duración promedio de las baterías sea menor a 1200 horas.")
```

```python
# ==============================================================================
# Problema #26: Prueba de Hipótesis para el Tiempo de Entrega
# ==============================================================================

# ------------------------------------------------------------------------------
# 1. Importación de Bibliotecas
# ------------------------------------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns

# ------------------------------------------------------------------------------
# 2. Definición de las Hipótesis y Parámetros
# ------------------------------------------------------------------------------
# El cliente sospecha que el tiempo es "mayor", lo que indica una prueba de una cola (superior).
# Hipótesis Nula (H₀): El tiempo promedio de entrega es de 25 minutos. (µ ≤ 25)
# Hipótesis Alternativa (Hₐ): El tiempo promedio de entrega es mayor a 25 minutos. (µ > 25)

# Parámetros del problema
media_poblacional_ho = 25  # Valor hipotetizado en H₀ (µ₀)
media_muestral = 28       # Media de la muestra (x̄)
desviacion_estandar_muestral = 4  # Desviación estándar de la muestra (s)
tamano_muestra = 15        # n
alpha = 0.05               # Nivel de significancia (α)

# Grados de libertad (df)
grados_libertad = tamano_muestra - 1

# ------------------------------------------------------------------------------
# 3. Cálculo del Estadístico t y el p-valor
# ------------------------------------------------------------------------------
# Fórmula del Estadístico t: t = (x̄ - µ₀) / (s / √n)
error_estandar = desviacion_estandar_muestral / np.sqrt(tamano_muestra)
t_calculado = (media_muestral - media_poblacional_ho) / error_estandar

# p-valor para una prueba de una cola (superior)
p_valor = stats.t.sf(t_calculado, df=grados_libertad)

# ------------------------------------------------------------------------------
# 4. Determinación del Valor Crítico y la Región de Rechazo
# ------------------------------------------------------------------------------
# Para una prueba de una cola (superior), el valor crítico t deja un área de α
# en la cola derecha de la distribución t de Student.
t_critico = stats.t.ppf(1 - alpha, df=grados_libertad)

# ------------------------------------------------------------------------------
# 5. Visualización del Resultado
# ------------------------------------------------------------------------------
# Configuración del gráfico
sns.set(style="whitegrid")
x = np.linspace(-4, 4, 1000)
y = stats.t.pdf(x, df=grados_libertad)

plt.figure(figsize=(12, 7))
plt.plot(x, y, color='blue', label=f'Distribución t de Student (df={grados_libertad})')

# Región de rechazo (área en la cola derecha)
plt.fill_between(x, y, where=(x > t_critico), color='red', alpha=0.5, label=f'Región de Rechazo (α = {alpha})')

# Líneas para el valor crítico y el estadístico calculado
plt.axvline(t_critico, color='red', linestyle='--', label=f'Valor Crítico t = {t_critico:.4f}')
plt.axvline(t_calculado, color='green', linestyle='-', label=f'Estadístico t Calculado = {t_calculado:.4f}')

# Etiquetas y título
plt.title('Prueba de Hipótesis para el Tiempo de Entrega (Una Cola Superior)')
plt.xlabel('Valor de t')
plt.ylabel('Densidad de Probabilidad')
plt.text(t_critico + 0.1, 0.05, 'Rechazar H₀', color='red', rotation=90)
plt.text(t_critico - 0.5, 0.05, 'No rechazar H₀', color='blue', rotation=90)
plt.legend()
plt.grid(True)
plt.show()

# ------------------------------------------------------------------------------
# 6. Conclusión
# ------------------------------------------------------------------------------
print("==============================================================================")
print("Resultados de la Prueba de Hipótesis (Distribución t de Student)")
print("==============================================================================")
print(f"Estadístico t calculado: {t_calculado:.4f}")
print(f"Grados de libertad (df): {grados_libertad}")
print(f"Valor crítico t: {t_critico:.4f}")
print(f"p-valor: {p_valor:.4f}")
print("------------------------------------------------------------------------------")

if p_valor < alpha:
    print(f"Decisión: Se rechaza la hipótesis nula (H₀).")
    print(f"El p-valor ({p_valor:.4f}) es menor que el nivel de significancia ({alpha}).")
    print(f"Conclusión: Con un nivel de significancia del 5%, existe evidencia estadística para respaldar la sospecha del cliente.")
    print("El tiempo promedio de entrega a domicilio es significativamente mayor a 25 minutos.")
else:
    print(f"Decisión: No se rechaza la hipótesis nula (H₀).")
    print(f"El p-valor ({p_valor:.4f}) es mayor o igual que el nivel de significancia ({alpha}).")
    print(f"Conclusión: No hay suficiente evidencia estadística para respaldar la sospecha del cliente.")
    print("No se puede concluir que el tiempo promedio de entrega sea mayor a 25 minutos.")
```

estudiantes tiene una media de 75 y s = 8. Con un nivel de significancia de α = 0,01.

*1.* Plantee las hipótesis nula y alternativa.

*2.* Calcule el estadístico t.

*3.* Determine los grados de libertad y el valor crítico, además de la región crítica y la región de
rechazo.

*4.* Concluya si se rechaza H0.


```python
# ==============================================================================
# Problema #27: Prueba de Hipótesis para la Media de una Clase de Matemáticas
# ==============================================================================

# ------------------------------------------------------------------------------
# 1. Importación de Bibliotecas
# ------------------------------------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns

# ------------------------------------------------------------------------------
# 2. Definición de las Hipótesis y Parámetros
# ------------------------------------------------------------------------------
# La afirmación es que el promedio es "diferente", lo que implica una prueba de dos colas.
# Hipótesis Nula (H₀): El promedio de la clase es igual a 80. (µ = 80)
# Hipótesis Alternativa (Hₐ): El promedio de la clase es diferente a 80. (µ ≠ 80)

# Parámetros del problema
media_poblacional_ho = 80  # Valor hipotetizado en H₀ (µ₀)
media_muestral = 75       # Media de la muestra (x̄)
desviacion_estandar_muestral = 8  # Desviación estándar de la muestra (s)
tamano_muestra = 12        # n
alpha = 0.01               # Nivel de significancia (α)

# Grados de libertad (df)
grados_libertad = tamano_muestra - 1

# ------------------------------------------------------------------------------
# 3. Cálculo del Estadístico t y el p-valor
# ------------------------------------------------------------------------------
# Fórmula del Estadístico t: t = (x̄ - µ₀) / (s / √n)
error_estandar = desviacion_estandar_muestral / np.sqrt(tamano_muestra)
t_calculado = (media_muestral - media_poblacional_ho) / error_estandar

# Para una prueba de dos colas, el p-valor se calcula como 2 * P(t > |t_calculado|)
p_valor = 2 * stats.t.sf(np.abs(t_calculado), df=grados_libertad)

# ------------------------------------------------------------------------------
# 4. Determinación del Valor Crítico y la Región de Rechazo
# ------------------------------------------------------------------------------
# Para una prueba de dos colas, el nivel de significancia se divide en dos: α/2.
alpha_dos_colas = alpha / 2
t_critico_inferior = stats.t.ppf(alpha_dos_colas, df=grados_libertad)
t_critico_superior = stats.t.ppf(1 - alpha_dos_colas, df=grados_libertad)

# ------------------------------------------------------------------------------
# 5. Visualización del Resultado
# ------------------------------------------------------------------------------
# Configuración del gráfico
sns.set(style="whitegrid")
x = np.linspace(-4, 4, 1000)
y = stats.t.pdf(x, df=grados_libertad)

plt.figure(figsize=(12, 7))
plt.plot(x, y, color='blue', label=f'Distribución t de Student (df={grados_libertad})')

# Región de rechazo (áreas en ambas colas)
plt.fill_between(x, y, where=(x < t_critico_inferior), color='red', alpha=0.5, label=f'Región de Rechazo (α/2 = {alpha_dos_colas})')
plt.fill_between(x, y, where=(x > t_critico_superior), color='red', alpha=0.5)

# Líneas para los valores críticos y el estadístico calculado
plt.axvline(t_critico_inferior, color='red', linestyle='--', label=f'Valores Críticos t = ±{t_critico_superior:.4f}')
plt.axvline(t_critico_superior, color='red', linestyle='--')
plt.axvline(t_calculado, color='green', linestyle='-', label=f'Estadístico t Calculado = {t_calculado:.4f}')

# Etiquetas y título
plt.title('Prueba de Hipótesis para el Promedio en Matemáticas (Dos Colas)')
plt.xlabel('Valor de t')
plt.ylabel('Densidad de Probabilidad')
plt.text(t_critico_inferior - 1.0, 0.05, 'Rechazar H₀', color='red', rotation=90)
plt.text(t_critico_superior + 0.1, 0.05, 'Rechazar H₀', color='red', rotation=90)
plt.text(0, 0.1, 'No rechazar H₀', color='blue')
plt.legend()
plt.grid(True)
plt.show()

# ------------------------------------------------------------------------------
# 6. Conclusión
# ------------------------------------------------------------------------------
print("==============================================================================")
print("Resultados de la Prueba de Hipótesis (Distribución t de Student)")
print("==============================================================================")
print(f"Estadístico t calculado: {t_calculado:.4f}")
print(f"Grados de libertad (df): {grados_libertad}")
print(f"Valores críticos t: ±{t_critico_superior:.4f}")
print(f"p-valor: {p_valor:.4f}")
print("------------------------------------------------------------------------------")

if p_valor < alpha:
    print(f"Decisión: Se rechaza la hipótesis nula (H₀).")
    print(f"El p-valor ({p_valor:.4f}) es menor que el nivel de significancia ({alpha}).")
    print(f"Conclusión: Con un nivel de significancia del 1%, existe evidencia estadística para afirmar que el promedio de la clase es significativamente diferente de 80 puntos.")
else:
    print(f"Decisión: No se rechaza la hipótesis nula (H₀).")
    print(f"El p-valor ({p_valor:.4f}) es mayor o igual que el nivel de significancia ({alpha}).")
    print(f"Conclusión: No hay suficiente evidencia estadística para afirmar que el promedio de la clase es diferente de 80 puntos.")
```

**31.** Una compañía de telefonía afirma que la varianza del tiempo de duración de las llamadas es de
σ2 = 2,5 minutos2. Se toma una muestra de n = 16 llamadas y se obtiene una varianza muestral
de s2 = 4. ¿Hay evidencia al 5 % de significancia de que la varianza poblacional difiere de 2.5?
Resuelva siguiendo el siguiente protocolo:

*1.* Plantee las hipótesis nula y alternativa.

*2.* Calcule el estadístico χ2.

*3.* Determine los grados de libertad y el valor crítico, además de la región crítica y la región de
rechazo.

*4.* Concluya si se rechaza H0.


```python
# ==============================================================================
# Problema #31: Prueba de Hipótesis para la Varianza del Tiempo de Llamadas
# ==============================================================================

# ------------------------------------------------------------------------------
# 1. Importación de Bibliotecas
# ------------------------------------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns

# ------------------------------------------------------------------------------
# 2. Definición de las Hipótesis y Parámetros
# ------------------------------------------------------------------------------
# La pregunta es si la varianza poblacional "difiere" de 2.5, lo que implica una prueba de dos colas.
# Hipótesis Nula (H₀): La varianza poblacional es igual a 2.5. (σ² = 2.5)
# Hipótesis Alternativa (Hₐ): La varianza poblacional es diferente de 2.5. (σ² ≠ 2.5)

# Parámetros del problema
varianza_poblacional_ho = 2.5  # Varianza hipotetizada en H₀ (σ²)
varianza_muestral = 4.0      # Varianza de la muestra (s²)
tamano_muestra = 16          # n
alpha = 0.05                 # Nivel de significancia (α)

# Grados de libertad (df)
grados_libertad = tamano_muestra - 1

# ------------------------------------------------------------------------------
# 3. Cálculo del Estadístico Chi-cuadrado y el p-valor
# ------------------------------------------------------------------------------
# Fórmula del Estadístico Chi-cuadrado (χ²): χ² = ((n - 1) * s²) / σ²
chi_cuadrado_calculado = ((tamano_muestra - 1) * varianza_muestral) / varianza_poblacional_ho

# Para una prueba de dos colas, el p-valor se calcula como 2 * P(χ² > χ²_calculado)
# o 2 * P(χ² < χ²_calculado), dependiendo del valor.
# Usamos el menor de los dos para la prueba de dos colas.
p_valor_inferior = stats.chi2.cdf(chi_cuadrado_calculado, df=grados_libertad)
p_valor_superior = stats.chi2.sf(chi_cuadrado_calculado, df=grados_libertad)
p_valor = 2 * min(p_valor_inferior, p_valor_superior)

# ------------------------------------------------------------------------------
# 4. Determinación de los Valores Críticos y la Región de Rechazo
# ------------------------------------------------------------------------------
# Para una prueba de dos colas, el nivel de significancia se divide en dos: α/2.
alpha_dos_colas = alpha / 2
chi_cuadrado_critico_inferior = stats.chi2.ppf(alpha_dos_colas, df=grados_libertad)
chi_cuadrado_critico_superior = stats.chi2.ppf(1 - alpha_dos_colas, df=grados_libertad)

# ------------------------------------------------------------------------------
# 5. Visualización del Resultado
# ------------------------------------------------------------------------------
# Configuración del gráfico
sns.set(style="whitegrid")
x = np.linspace(0, 40, 500)
y = stats.chi2.pdf(x, df=grados_libertad)

plt.figure(figsize=(12, 7))
plt.plot(x, y, color='blue', label=f'Distribución Chi-cuadrado (df={grados_libertad})')

# Región de rechazo (áreas en ambas colas)
plt.fill_between(x, y, where=(x < chi_cuadrado_critico_inferior), color='red', alpha=0.5, label='Región de Rechazo (α/2)')
plt.fill_between(x, y, where=(x > chi_cuadrado_critico_superior), color='red', alpha=0.5)

# Líneas para los valores críticos y el estadístico calculado
plt.axvline(chi_cuadrado_critico_inferior, color='red', linestyle='--', label=f'Valor Crítico Inferior = {chi_cuadrado_critico_inferior:.4f}')
plt.axvline(chi_cuadrado_critico_superior, color='red', linestyle='--', label=f'Valor Crítico Superior = {chi_cuadrado_critico_superior:.4f}')
plt.axvline(chi_cuadrado_calculado, color='green', linestyle='-', label=f'Estadístico χ² Calculado = {chi_cuadrado_calculado:.4f}')

# Etiquetas y título
plt.title('Prueba de Hipótesis para la Varianza (Dos Colas)')
plt.xlabel('Valor de Chi-cuadrado (χ²)')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()

# ------------------------------------------------------------------------------
# 6. Conclusión
# ------------------------------------------------------------------------------
print("==============================================================================")
print("Resultados de la Prueba de Hipótesis (Distribución Chi-cuadrado)")
print("==============================================================================")
print(f"Estadístico χ² calculado: {chi_cuadrado_calculado:.4f}")
print(f"Grados de libertad (df): {grados_libertad}")
print(f"Valores críticos χ²: {chi_cuadrado_critico_inferior:.4f} y {chi_cuadrado_critico_superior:.4f}")
print(f"p-valor: {p_valor:.4f}")
print("------------------------------------------------------------------------------")

if p_valor < alpha:
    print(f"Decisión: Se rechaza la hipótesis nula (H₀).")
    print(f"El p-valor ({p_valor:.4f}) es menor que el nivel de significancia ({alpha}).")
    print(f"Conclusión: Con un nivel de significancia del 5%, existe evidencia estadística para afirmar que la varianza poblacional de la duración de las llamadas es significativamente diferente de 2.5 minutos².")
else:
    print(f"Decisión: No se rechaza la hipótesis nula (H₀).")
    print(f"El p-valor ({p_valor:.4f}) es mayor o igual que el nivel de significancia ({alpha}).")
    print(f"Conclusión: No hay suficiente evidencia estadística para afirmar que la varianza poblacional de la duración de las llamadas sea diferente de 2.5 minutos².")
```

```python
# ==============================================================================
# Problema #32: Prueba de Hipótesis para la Varianza del Llenado de Botellas
# ==============================================================================

# ------------------------------------------------------------------------------
# 1. Importación de Bibliotecas
# ------------------------------------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns

# ------------------------------------------------------------------------------
# 2. Definición de las Hipótesis y Parámetros
# ------------------------------------------------------------------------------
# La sospecha es que la varianza es "mayor", lo que implica una prueba de una cola (superior).
# Hipótesis Nula (H₀): La varianza poblacional es igual o menor a 1.2. (σ² ≤ 1.2)
# Hipótesis Alternativa (Hₐ): La varianza poblacional es mayor a 1.2. (σ² > 1.2)

# Parámetros del problema
varianza_poblacional_ho = 1.2  # Varianza hipotetizada en H₀ (σ²)
varianza_muestral = 2.0      # Varianza de la muestra (s²)
tamano_muestra = 20          # n
alpha = 0.01                 # Nivel de significancia (α)

# Grados de libertad (df)
grados_libertad = tamano_muestra - 1

# ------------------------------------------------------------------------------
# 3. Cálculo del Estadístico Chi-cuadrado y el p-valor
# ------------------------------------------------------------------------------
# Fórmula del Estadístico Chi-cuadrado (χ²): χ² = ((n - 1) * s²) / σ²
chi_cuadrado_calculado = ((tamano_muestra - 1) * varianza_muestral) / varianza_poblacional_ho

# Para una prueba de una cola (superior), el p-valor es el área a la derecha del estadístico.
p_valor = stats.chi2.sf(chi_cuadrado_calculado, df=grados_libertad)

# ------------------------------------------------------------------------------
# 4. Determinación del Valor Crítico y la Región de Rechazo
# ------------------------------------------------------------------------------
# Para una prueba de una cola (superior), el valor crítico χ² deja un área de α
# en la cola derecha de la distribución Chi-cuadrado.
chi_cuadrado_critico = stats.chi2.ppf(1 - alpha, df=grados_libertad)

# ------------------------------------------------------------------------------
# 5. Visualización del Resultado
# ------------------------------------------------------------------------------
# Configuración del gráfico
sns.set(style="whitegrid")
x = np.linspace(0, 40, 500)
y = stats.chi2.pdf(x, df=grados_libertad)

plt.figure(figsize=(12, 7))
plt.plot(x, y, color='blue', label=f'Distribución Chi-cuadrado (df={grados_libertad})')

# Región de rechazo (área en la cola derecha)
plt.fill_between(x, y, where=(x > chi_cuadrado_critico), color='red', alpha=0.5, label=f'Región de Rechazo (α = {alpha})')

# Líneas para el valor crítico y el estadístico calculado
plt.axvline(chi_cuadrado_critico, color='red', linestyle='--', label=f'Valor Crítico χ² = {chi_cuadrado_critico:.4f}')
plt.axvline(chi_cuadrado_calculado, color='green', linestyle='-', label=f'Estadístico χ² Calculado = {chi_cuadrado_calculado:.4f}')

# Etiquetas y título
plt.title('Prueba de Hipótesis para la Varianza (Una Cola Superior)')
plt.xlabel('Valor de Chi-cuadrado (χ²)')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()

# ------------------------------------------------------------------------------
# 6. Conclusión
# ------------------------------------------------------------------------------
print("==============================================================================")
print("Resultados de la Prueba de Hipótesis (Distribución Chi-cuadrado)")
print("==============================================================================")
print(f"Estadístico χ² calculado: {chi_cuadrado_calculado:.4f}")
print(f"Grados de libertad (df): {grados_libertad}")
print(f"Valor crítico χ²: {chi_cuadrado_critico:.4f}")
print(f"p-valor: {p_valor:.4f}")
print("------------------------------------------------------------------------------")

if p_valor < alpha:
    print(f"Decisión: Se rechaza la hipótesis nula (H₀).")
    print(f"El p-valor ({p_valor:.4f}) es menor que el nivel de significancia ({alpha}).")
    print(f"Conclusión: Con un nivel de significancia del 1%, existe evidencia estadística para afirmar que la varianza poblacional del llenado es significativamente mayor a 1.2 ml².")
else:
    print(f"Decisión: No se rechaza la hipótesis nula (H₀).")
    print(f"El p-valor ({p_valor:.4f}) es mayor o igual que el nivel de significancia ({alpha}).")
    print(f"Conclusión: No hay suficiente evidencia estadística para afirmar que la varianza poblacional del llenado sea mayor a 1.2 ml².")
```

**33.** El control de calidad de una fábrica sostiene que la varianza en el grosor de piezas metálicas es de
al menos σ2 = 0,5 mm2. Una muestra de n = 12 piezas dio una varianza s2 = 0,3. ¿Se puede
concluir al 5 % que la varianza poblacional es menor que la garantizada?
Resuelva siguiendo el siguiente protocolo:

*1.* Plantee las hipótesis nula y alternativa.

*2.* texto en cursiva Calcule el estadístico χ2.

*3.* Determine los grados de libertad y el valor
crítico, además de la región crítica y la región de
rechazo.

*4.* Concluya si se rechaza H0.


```python
# ==============================================================================
# Problema #33: Prueba de Hipótesis para la Varianza del Grosor de Piezas Metálicas
# ==============================================================================

# ------------------------------------------------------------------------------
# 1. Importación de Bibliotecas
# ------------------------------------------------------------------------------
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns

# ------------------------------------------------------------------------------
# 2. Definición de las Hipótesis y Parámetros
# ------------------------------------------------------------------------------
# La pregunta es si la varianza es "menor" que la garantizada, lo que implica una prueba de una cola (inferior).
# Hipótesis Nula (H₀): La varianza poblacional es igual o mayor a 0.5. (σ² ≥ 0.5)
# Hipótesis Alternativa (Hₐ): La varianza poblacional es menor a 0.5. (σ² < 0.5)

# Parámetros del problema
varianza_poblacional_ho = 0.5  # Varianza hipotetizada en H₀ (σ²)
varianza_muestral = 0.3      # Varianza de la muestra (s²)
tamano_muestra = 12          # n
alpha = 0.05                 # Nivel de significancia (α)

# Grados de libertad (df)
grados_libertad = tamano_muestra - 1

# ------------------------------------------------------------------------------
# 3. Cálculo del Estadístico Chi-cuadrado y el p-valor
# ------------------------------------------------------------------------------
# Fórmula del Estadístico Chi-cuadrado (χ²): χ² = ((n - 1) * s²) / σ²
chi_cuadrado_calculado = ((tamano_muestra - 1) * varianza_muestral) / varianza_poblacional_ho

# Para una prueba de una cola (inferior), el p-valor es el área a la izquierda del estadístico.
p_valor = stats.chi2.cdf(chi_cuadrado_calculado, df=grados_libertad)

# ------------------------------------------------------------------------------
# 4. Determinación del Valor Crítico y la Región de Rechazo
# ------------------------------------------------------------------------------
# Para una prueba de una cola (inferior), el valor crítico χ² deja un área de α
# en la cola izquierda de la distribución Chi-cuadrado.
chi_cuadrado_critico = stats.chi2.ppf(alpha, df=grados_libertad)

# ------------------------------------------------------------------------------
# 5. Visualización del Resultado
# ------------------------------------------------------------------------------
# Configuración del gráfico
sns.set(style="whitegrid")
x = np.linspace(0, 30, 500)
y = stats.chi2.pdf(x, df=grados_libertad)

plt.figure(figsize=(12, 7))
plt.plot(x, y, color='blue', label=f'Distribución Chi-cuadrado (df={grados_libertad})')

# Región de rechazo (área en la cola izquierda)
plt.fill_between(x, y, where=(x < chi_cuadrado_critico), color='red', alpha=0.5, label=f'Región de Rechazo (α = {alpha})')

# Líneas para el valor crítico y el estadístico calculado
plt.axvline(chi_cuadrado_critico, color='red', linestyle='--', label=f'Valor Crítico χ² = {chi_cuadrado_critico:.4f}')
plt.axvline(chi_cuadrado_calculado, color='green', linestyle='-', label=f'Estadístico χ² Calculado = {chi_cuadrado_calculado:.4f}')

# Etiquetas y título
plt.title('Prueba de Hipótesis para la Varianza del Grosor (Una Cola Inferior)')
plt.xlabel('Valor de Chi-cuadrado (χ²)')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True)
plt.show()

# ------------------------------------------------------------------------------
# 6
```
