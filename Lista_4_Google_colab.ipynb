1. Se desea estimar la altura promedio de estudiantes universitarios. En una muestra de 50 estudiantes,
se obtuvo una media de 170 cm. Se sabe que la desviación estándar poblacional es de 10 cm.
Construye un intervalo de confianza del 95 % para la media poblacional.

```python
# ==============================================================================
# EJERCICIO 1: Altura Promedio de Estudiantes Universitarios
# Intervalo de Confianza del 95% (IC Z)
# ==============================================================================

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def calcular_y_graficar_ic_z(x_bar, sigma, n, nivel_confianza, titulo, unidad):
    """Calcula, imprime y grafica el Intervalo de Confianza Z para la media poblacional."""
    alpha = 1 - nivel_confianza
    error_estandar = sigma / np.sqrt(n)
    z_critico = stats.norm.ppf(1 - alpha / 2)
    margen_error = z_critico * error_estandar
    limite_inferior = x_bar - margen_error
    limite_superior = x_bar + margen_error

    print(f"\n--- EJERCICIO: {titulo} ---")
    print(f"IC {nivel_confianza*100:.0f}%: [{limite_inferior:.4f}, {limite_superior:.4f}] {unidad}")
    print(f"Valor Z Crítico: $\\pm{z_critico:.4f}")

    # Gráfica
    rango = np.linspace(x_bar - 4 * error_estandar, x_bar + 4 * error_estandar, 1000)
    pdf = stats.norm.pdf(rango, loc=x_bar, scale=error_estandar)

    plt.figure(figsize=(10, 6))
    plt.plot(rango, pdf, color='gray', label='Distribución Muestral')
    plt.fill_between(rango, 0, pdf,
                     where=(rango >= limite_inferior) & (rango <= limite_superior),
                     color='green', alpha=0.3, label=f'Región de Confianza ({nivel_confianza*100:.0f}%)')
    plt.axvline(limite_inferior, color='green', linestyle='--', linewidth=2, label=f'Límite Inferior ({limite_inferior:.2f})')
    plt.axvline(limite_superior, color='green', linestyle='--', linewidth=2, label=f'Límite Superior ({limite_superior:.2f})')
    plt.axvline(x_bar, color='blue', linestyle='-', linewidth=2, label=f'Media Muestral ({x_bar})')
    plt.title(f'Intervalo de Confianza del {nivel_confianza*100:.0f}% para la Media de {titulo}')
    plt.xlabel(f'Valor ({unidad})')
    plt.ylabel('Densidad de Probabilidad')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

    print(f"CONCLUSIÓN PRÁCTICA: Con un {nivel_confianza*100:.0f}% de confianza, la altura promedio real ($\mu$) está entre {limite_inferior:.2f} y {limite_superior:.2f} {unidad}.")
    print("----------------------------------------------------------\n")

# DATOS DEL EJERCICIO 1
calcular_y_graficar_ic_z(
    x_bar=170.0,
    sigma=10.0,
    n=50,
    nivel_confianza=0.95,
    titulo="Altura Promedio de Estudiantes",
    unidad="cm"
)
```

6. Una cadena de cafeterías desea estimar la temperatura promedio (°C) del café que sirve a sus clien-
tes. Se tomó una muestra aleatoria de 12 tazas donde la media fue de 84°C, y la desviación estándar
5°C. Con un nivel de confianza del 95 % calcule el intervalo de confianza para la temperatura media
poblacional del café.

```python
# ==============================================================================
# EJERCICIO 6: Temperatura Promedio del Café
# Intervalo de Confianza del 95% (IC t de Student)
# n < 30 (n=12) y sigma desconocida (usamos s)
# ==============================================================================

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def calcular_y_graficar_ic_t(x_bar, s, n, nivel_confianza, titulo, unidad):
    """Calcula, imprime y grafica el Intervalo de Confianza t de Student para la media poblacional."""
    alpha = 1 - nivel_confianza
    df = n - 1 # Grados de libertad (degrees of freedom)

    # 1. CÁLCULO DEL ERROR ESTÁNDAR
    # Fórmula: SE = s / sqrt(n)
    error_estandar = s / np.sqrt(n)

    # 2. CÁLCULO DEL VALOR t CRÍTICO
    # stats.t.ppf(1 - alpha/2, df)
    t_critico = stats.t.ppf(1 - alpha / 2, df)

    # 3. CÁLCULO DEL MARGEN DE ERROR
    # Fórmula: ME = t_critico * SE
    margen_error = t_critico * error_estandar

    # 4. CÁLCULO DE LOS LÍMITES DEL INTERVALO
    limite_inferior = x_bar - margen_error
    limite_superior = x_bar + margen_error

    print(f"\n--- EJERCICIO: {titulo} ---")
    print(f"Estadístico de Prueba: t de Student (df={df})")
    print(f"IC {nivel_confianza*100:.0f}%: [{limite_inferior:.4f}, {limite_superior:.4f}] {unidad}")
    print(f"Valor t Crítico: $\\pm{t_critico:.4f}")

    # Gráfica
    # Rango de 4 errores estándar alrededor de la media muestral
    rango = np.linspace(x_bar - 4 * error_estandar, x_bar + 4 * error_estandar, 1000)
    # Función de densidad de probabilidad (PDF) para la distribución t
    pdf = stats.t.pdf( (rango - x_bar) / error_estandar, df=df) / error_estandar

    plt.figure(figsize=(10, 6))
    plt.plot(rango, pdf, color='gray', label=f'Distribución t de Student (df={df})')

    # Rellenar la región de confianza
    plt.fill_between(rango, 0, pdf,
                     where=(rango >= limite_inferior) & (rango <= limite_superior),
                     color='green', alpha=0.3, label=f'Región de Confianza ({nivel_confianza*100:.0f}%)')

    # Marcar límites y media
    plt.axvline(limite_inferior, color='green', linestyle='--', linewidth=2, label=f'Límite Inferior ({limite_inferior:.2f})')
    plt.axvline(limite_superior, color='green', linestyle='--', linewidth=2, label=f'Límite Superior ({limite_superior:.2f})')
    plt.axvline(x_bar, color='blue', linestyle='-', linewidth=2, label=f'Media Muestral ({x_bar})')

    plt.title(f'Intervalo de Confianza del {nivel_confianza*100:.0f}% para la Media de {titulo}')
    plt.xlabel(f'Valor ({unidad})')
    plt.ylabel('Densidad de Probabilidad')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

    print(f"CONCLUSIÓN PRÁCTICA: Con un {nivel_confianza*100:.0f}% de confianza, la temperatura promedio real ($\mu$) del café está entre {limite_inferior:.2f} y {limite_superior:.2f} {unidad}.")
    print("----------------------------------------------------------\n")

# DATOS DEL EJERCICIO 6
calcular_y_graficar_ic_t(
    x_bar=84.0,
    s=5.0,
    n=12,
    nivel_confianza=0.95,
    titulo="Temperatura Promedio del Café",
    unidad="°C"
)
```

```python
# ==============================================================================
# EJERCICIO 8: Altura Promedio de Plántulas de Maíz
# Intervalo de Confianza del 99% (IC t de Student)
# n < 30 (n=10) y sigma desconocida (usamos s)
# ==============================================================================

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def calcular_y_graficar_ic_t(x_bar, s, n, nivel_confianza, titulo, unidad):
    """Calcula, imprime y grafica el Intervalo de Confianza t de Student para la media poblacional."""
    alpha = 1 - nivel_confianza
    df = n - 1 # Grados de libertad (degrees of freedom)

    # 1. CÁLCULO DEL ERROR ESTÁNDAR
    error_estandar = s / np.sqrt(n)

    # 2. CÁLCULO DEL VALOR t CRÍTICO
    t_critico = stats.t.ppf(1 - alpha / 2, df)

    # 3. CÁLCULO DEL MARGEN DE ERROR
    margen_error = t_critico * error_estandar

    # 4. CÁLCULO DE LOS LÍMITES DEL INTERVALO
    limite_inferior = x_bar - margen_error
    limite_superior = x_bar + margen_error

    print(f"\n--- EJERCICIO: {titulo} ---")
    print(f"Estadístico de Prueba: t de Student (df={df})")
    print(f"IC {nivel_confianza*100:.0f}%: [{limite_inferior:.4f}, {limite_superior:.4f}] {unidad}")
    print(f"Valor t Crítico: $\\pm{t_critico:.4f}")

    # Gráfica
    rango = np.linspace(x_bar - 4 * error_estandar, x_bar + 4 * error_estandar, 1000)
    pdf = stats.t.pdf( (rango - x_bar) / error_estandar, df=df) / error_estandar

    plt.figure(figsize=(10, 6))
    plt.plot(rango, pdf, color='gray', label=f'Distribución t de Student (df={df})')

    plt.fill_between(rango, 0, pdf,
                     where=(rango >= limite_inferior) & (rango <= limite_superior),
                     color='orange', alpha=0.3, label=f'Región de Confianza ({nivel_confianza*100:.0f}%)')

    plt.axvline(limite_inferior, color='orange', linestyle='--', linewidth=2, label=f'Límite Inferior ({limite_inferior:.2f})')
    plt.axvline(limite_superior, color='orange', linestyle='--', linewidth=2, label=f'Límite Superior ({limite_superior:.2f})')
    plt.axvline(x_bar, color='blue', linestyle='-', linewidth=2, label=f'Media Muestral ({x_bar})')

    plt.title(f'Intervalo de Confianza del {nivel_confianza*100:.0f}% para la Media de {titulo}')
    plt.xlabel(f'Valor ({unidad})')
    plt.ylabel('Densidad de Probabilidad')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

    print(f"CONCLUSIÓN PRÁCTICA: Con un {nivel_confianza*100:.0f}% de confianza, la altura promedio real ($\mu$) de las plántulas está entre {limite_inferior:.2f} y {limite_superior:.2f} {unidad}.")
    print("----------------------------------------------------------\n")

# DATOS DEL EJERCICIO 8
calcular_y_graficar_ic_t(
    x_bar=25.0,
    s=3.0,
    n=10,
    nivel_confianza=0.99,
    titulo="Altura Promedio de Plántulas de Maíz",
    unidad="cm"
)
```

11. El departamento de salud planea una encuesta para estimar la proporción de residentes vacunados
contra la influenza. En un estudio piloto, el 72 % de una muestra mostró estar vacunado. Determine
el tamaño de muestra necesario para estimar esta proporción con un margen de error del 3 % y un
nivel de confianza del 95 %. Si no se tuviera el estudio piloto, ¿cuál sería el tamaño requerido?

```python
# ==============================================================================
# EJERCICIO 11: Tamaño de Muestra para Proporción de Residentes Vacunados
# Fórmulas: n = [Z^2 * p * (1-p)] / E^2
# ==============================================================================

import numpy as np
import scipy.stats as stats
import math
import matplotlib.pyplot as plt

def graficar_z_critico(z_critico, nivel_confianza, titulo):
    """Grafica la distribución Normal Estándar, destacando la región de confianza
       definida por el Z crítico."""

    rango = np.linspace(-4, 4, 1000)
    pdf = stats.norm.pdf(rango)

    plt.figure(figsize=(10, 6))
    plt.plot(rango, pdf, color='black', label='Distribución Normal Estándar (Z)')

    # Rellenar la región central de confianza
    limite_inf = -z_critico
    limite_sup = z_critico
    rango_confianza = rango[(rango >= limite_inf) & (rango <= limite_sup)]

    plt.fill_between(rango_confianza, stats.norm.pdf(rango_confianza), color='blue', alpha=0.3,
                     label=f'Región de Confianza ({nivel_confianza*100:.0f}%)')

    # Marcar valores críticos Z
    plt.axvline(limite_inf, color='red', linestyle='--', label=f'Valor Crítico Z = {limite_inf:.4f}')
    plt.axvline(limite_sup, color='red', linestyle='--')

    plt.title(f'Gráfica del Valor Crítico Z para {titulo}')
    plt.xlabel('Estadístico Z')
    plt.ylabel('Densidad de Probabilidad')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

# --- DATOS DEL PROBLEMA ---
NIVEL_CONFIANZA = 0.95
MARGEN_ERROR_E = 0.03 # 3%
ALPHA = 1 - NIVEL_CONFIANZA

print("--- EJERCICIO 11: Tamaño de Muestra para Proporción de Vacunados ---")

# 1. VALOR Z CRÍTICO (Para 95% de confianza)
Z_CRITICO = stats.norm.ppf(1 - ALPHA / 2)
print(f"Valor Z Crítico ($Z_{{\\alpha/2}}$ para 95%): {Z_CRITICO:.4f}")

# A. CÁLCULO USANDO EL ESTUDIO PILOTO (p = 0.72)
P_PILOTO = 0.72
Q_PILOTO = 1 - P_PILOTO
n_calculado_piloto = (Z_CRITICO**2 * P_PILOTO * Q_PILOTO) / (MARGEN_ERROR_E**2)
n_requerido_piloto = math.ceil(n_calculado_piloto)
print(f"\n--- A. CON ESTUDIO PILOTO (p = 0.72) ---\nn calculado: {n_calculado_piloto:.2f}\nTamaño de muestra REQUERIDO: {n_requerido_piloto}")

# B. CÁLCULO SIN ESTUDIO PILOTO (p = 0.50)
P_CONSERVADOR = 0.50
Q_CONSERVADOR = 1 - P_CONSERVADOR
n_calculado_conservador = (Z_CRITICO**2 * P_CONSERVADOR * Q_CONSERVADOR) / (MARGEN_ERROR_E**2)
n_requerido_conservador = math.ceil(n_calculado_conservador)
print(f"\n--- B. SIN ESTUDIO PILOTO (p = 0.50) ---\nn calculado: {n_calculado_conservador:.2f}\nTamaño de muestra REQUERIDO: {n_requerido_conservador}")

# 6. Gráfica (Implementación para Z crítico)
graficar_z_critico(Z_CRITICO, NIVEL_CONFIANZA, "Tamaño de Muestra para Proporción (95% IC)")

print("\nCONCLUSIÓN PRÁCTICA:")
print(f"Usando el estudio piloto (p=0.72), se requieren {n_requerido_piloto} residentes.")
print(f"Si no hay datos, se requiere el tamaño máximo de {n_requerido_conservador} (usando p=0.5).")
print("----------------------------------------------------------\n")
```

```python
# ==============================================================================
# EJERCICIO 12: Tamaño de Muestra para Tasa de Devoluciones
# Fórmulas: n = [Z^2 * p * (1-p)] / E^2
# ==============================================================================

import numpy as np
import scipy.stats as stats
import math
import matplotlib.pyplot as plt

def graficar_z_critico(z_critico, nivel_confianza, titulo):
    """Grafica la distribución Normal Estándar, destacando la región de confianza
       definida por el Z crítico."""

    rango = np.linspace(-4, 4, 1000)
    pdf = stats.norm.pdf(rango)

    plt.figure(figsize=(10, 6))
    plt.plot(rango, pdf, color='black', label='Distribución Normal Estándar (Z)')

    limite_inf = -z_critico
    limite_sup = z_critico
    rango_confianza = rango[(rango >= limite_inf) & (rango <= limite_sup)]

    plt.fill_between(rango_confianza, stats.norm.pdf(rango_confianza), color='blue', alpha=0.3,
                     label=f'Región de Confianza ({nivel_confianza*100:.0f}%)')

    plt.axvline(limite_inf, color='red', linestyle='--', label=f'Valor Crítico Z = {limite_inf:.4f}')
    plt.axvline(limite_sup, color='red', linestyle='--')

    plt.title(f'Gráfica del Valor Crítico Z para {titulo}')
    plt.xlabel('Estadístico Z')
    plt.ylabel('Densidad de Probabilidad')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

# --- DATOS DEL PROBLEMA ---
NIVEL_CONFIANZA = 0.90
MARGEN_ERROR_E = 0.04 # 4%
P_PREVIO = 0.07       # 7%
Q_PREVIO = 1 - P_PREVIO
ALPHA = 1 - NIVEL_CONFIANZA

print("--- EJERCICIO 12: Tamaño de Muestra para Tasa de Devoluciones ---")

# 1. VALOR Z CRÍTICO (Para 90% de confianza)
Z_CRITICO = stats.norm.ppf(1 - ALPHA / 2)
print(f"Valor Z Crítico ($Z_{{\\alpha/2}}$ para 90%): {Z_CRITICO:.4f}")

# 2. CÁLCULO USANDO DATOS HISTÓRICOS
n_calculado = (Z_CRITICO**2 * P_PREVIO * Q_PREVIO) / (MARGEN_ERROR_E**2)
n_requerido = math.ceil(n_calculado)

print("\n--- CÁLCULO CON DATOS PREVIOS (p = 0.07) ---")
print(f"p * (1-p) = {P_PREVIO * Q_PREVIO:.4f}")
print(f"n calculado: {n_calculado:.2f}")
print(f"Tamaño de muestra REQUERIDO: {n_requerido}")

# 6. Gráfica (Implementación para Z crítico)
graficar_z_critico(Z_CRITICO, NIVEL_CONFIANZA, "Tamaño de Muestra para Proporción (90% IC)")

print("\nCONCLUSIÓN PRÁCTICA:")
print(f"Para estimar la proporción de devoluciones con un Margen de Error de 4% y 90% de confianza, se necesitan {n_requerido} pedidos en la muestra, utilizando la estimación previa de p=0.07.")
print("----------------------------------------------------------\n")
```

```python
# ==============================================================================
# EJERCICIO 14: Tamaño de Muestra para Proporción de Apoyo a Política
# Fórmulas: n = [Z^2 * p * (1-p)] / E^2
# ==============================================================================

import numpy as np
import scipy.stats as stats
import math
import matplotlib.pyplot as plt

def graficar_z_critico(z_critico, nivel_confianza, titulo):
    """Grafica la distribución Normal Estándar, destacando la región de confianza
       definida por el Z crítico."""

    rango = np.linspace(-4, 4, 1000)
    pdf = stats.norm.pdf(rango)

    plt.figure(figsize=(10, 6))
    plt.plot(rango, pdf, color='black', label='Distribución Normal Estándar (Z)')

    limite_inf = -z_critico
    limite_sup = z_critico
    rango_confianza = rango[(rango >= limite_inf) & (rango <= limite_sup)]

    plt.fill_between(rango_confianza, stats.norm.pdf(rango_confianza), color='blue', alpha=0.3,
                     label=f'Región de Confianza ({nivel_confianza*100:.0f}%)')

    plt.axvline(limite_inf, color='red', linestyle='--', label=f'Valor Crítico Z = {limite_inf:.4f}')
    plt.axvline(limite_sup, color='red', linestyle='--')

    plt.title(f'Gráfica del Valor Crítico Z para {titulo}')
    plt.xlabel('Estadístico Z')
    plt.ylabel('Densidad de Probabilidad')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

# --- DATOS DEL PROBLEMA ---
NIVEL_CONFIANZA = 0.95
MARGEN_ERROR_E = 0.05 # 5%
P_PREVIO = 0.56       # 56%
ALPHA = 1 - NIVEL_CONFIANZA

print("--- EJERCICIO 14: Tamaño de Muestra para Proporción de Apoyo ---")

# 1. VALOR Z CRÍTICO (Para 95% de confianza)
Z_CRITICO = stats.norm.ppf(1 - ALPHA / 2)
print(f"Valor Z Crítico ($Z_{{\\alpha/2}}$ para 95%): {Z_CRITICO:.4f}")

# A. CÁLCULO USANDO DATOS PREVIOS (p = 0.56)
P_PREVIO = 0.56
Q_PREVIO = 1 - P_PREVIO
n_calculado_previo = (Z_CRITICO**2 * P_PREVIO * Q_PREVIO) / (MARGEN_ERROR_E**2)
n_requerido_previo = math.ceil(n_calculado_previo)
print(f"\n--- CÁLCULO CON DATOS PREVIOS (p = 0.56) ---\nn calculado: {n_calculado_previo:.2f}\nTamaño de muestra REQUERIDO: {n_requerido_previo}")

# B. CÁLCULO CONSERVADOR (p = 0.50)
P_CONSERVADOR = 0.50
Q_CONSERVADOR = 0.50
n_calculado_conservador = (Z_CRITICO**2 * P_CONSERVADOR * Q_CONSERVADOR) / (MARGEN_ERROR_E**2)
n_requerido_conservador = math.ceil(n_calculado_conservador)

# 6. Gráfica (Implementación para Z crítico)
graficar_z_critico(Z_CRITICO, NIVEL_CONFIANZA, "Tamaño de Muestra para Proporción (95% IC)")

print("\n--- JUSTIFICACIÓN DEL USO DE p = 0.5 ---")
print("Si no hay datos preliminares, se usa $p=0.5$ porque el producto $p(1-p)$ alcanza su valor máximo de $0.25$ cuando $p=0.5$.")
print("Al maximizar el numerador de la fórmula, se obtiene el **tamaño de muestra más grande** posible ({}) que garantiza que el margen de error deseado se cumpla, independientemente de cuál sea la proporción real.".format(n_requerido_conservador))

print("\nCONCLUSIÓN PRÁCTICA:")
print(f"Para estimar la proporción de apoyo con un Margen de Error de 5% y 95% de confianza, se requieren {n_requerido_previo} padres, utilizando la estimación informal de p=0.56.")
print("----------------------------------------------------------\n")
```

media de 495 g. La desviación estándar poblacional es σ = 15 g. Con un nivel de de significancia
α = 0,05.

*   Plantee las hipótesis nula y alternativa.
*   Calcule el estadístico Z.
*   Determine la región crítica y la región de rechazo.
*   Concluya si se rechaza H0.





```python
# ==============================================================================
# EJERCICIO 21: Prueba de Hipótesis Z para la Media (Unilateral Izquierda)
# ==============================================================================

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# --- DATOS DEL PROBLEMA ---
MU_0 = 500.0   # Peso promedio afirmado (g)
X_BAR = 495.0  # Media muestral (g)
SIGMA = 15.0   # Desviación estándar poblacional (g)
N = 40         # Tamaño de la muestra
ALPHA = 0.05   # Nivel de significancia (5%)

print("--- EJERCICIO 21: Prueba de Hipótesis Z para la Media (Cereal) ---")

# 1. PLANTÉE LAS HIPÓTESIS NULA Y ALTERNATIVA
print("\n1. Hipótesis:")
print(f"   H0 (Hipótesis Nula): La máquina está calibrada (μ ≥ {MU_0} g)")
print(f"   H1 (Hipótesis Alternativa): La máquina está subllenando (μ < {MU_0} g) [Prueba Unilateral Izquierda]")

# 2. CALCULE EL ESTADÍSTICO Z
# Error Estándar (SE): SIGMA / sqrt(N)
SE = SIGMA / np.sqrt(N)
# Estadístico de Prueba Z: (X_BAR - MU_0) / SE
Z_ESTADISTICO = (X_BAR - MU_0) / SE
print(f"\n2. Estadístico Z Calculado:")
print(f"   Error Estándar (SE): {SE:.4f} g")
print(f"   Z Calculado: {Z_ESTADISTICO:.4f}")

# 3. DETERMINE LA REGIÓN CRÍTICA Y LA REGIÓN DE RECHAZO
# Para prueba unilateral izquierda con α = 0.05, buscamos Z_α
Z_CRITICO = stats.norm.ppf(ALPHA) # stats.norm.ppf(0.05)
print(f"\n3. Región Crítica (α = {ALPHA}):")
print(f"   Valor Crítico Z (Z_α): {Z_CRITICO:.4f}")
print(f"   Región de Rechazo (RR): Z < {Z_CRITICO:.4f}")

# 4. CONCLUYA SI SE RECHAZA H0
print("\n4. Decisión y Conclusión:")
if Z_ESTADISTICO < Z_CRITICO:
    DECISION = "RECHAZAR H0"
    CONCLUSION = "Existe evidencia suficiente (al nivel α=0.05) para concluir que el peso promedio de los paquetes es menor a 500 g, confirmando la sospecha de subllenado."
else:
    DECISION = "NO RECHAZAR H0"
    CONCLUSION = "No existe evidencia suficiente (al nivel α=0.05) para concluir que el peso promedio de los paquetes es menor a 500 g."

print(f"   Decisión: {DECISION} (Dado que {Z_ESTADISTICO:.4f} es {'< (menor que)' if Z_ESTADISTICO < Z_CRITICO else '≥ (mayor o igual a)'} {Z_CRITICO:.4f})")
print(f"   Conclusión: {CONCLUSION}")

# 6. GRÁFICA DE LA PRUEBA DE HIPÓTESIS
rango = np.linspace(-4, 4, 1000)
pdf = stats.norm.pdf(rango)

plt.figure(figsize=(10, 6))
plt.plot(rango, pdf, color='black', label='Distribución Normal Estándar (Z)')

# Rellenar Región de Rechazo (cola izquierda)
rango_rechazo = rango[rango < Z_CRITICO]
plt.fill_between(rango_rechazo, stats.norm.pdf(rango_rechazo), color='red', alpha=0.3, label=f'Región de Rechazo (α={ALPHA})')

# Marcar valor crítico Z
plt.axvline(Z_CRITICO, color='red', linestyle='--', linewidth=2, label=f'Z Crítico = {Z_CRITICO:.4f}')

# Marcar estadístico de prueba
plt.axvline(Z_ESTADISTICO, color='green', linestyle='-', linewidth=3, label=f'Z Calculado = {Z_ESTADISTICO:.4f}')

plt.title('Prueba de Hipótesis Z Unilateral Izquierda (α=0.05)')
plt.xlabel('Estadístico Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

print("----------------------------------------------------------\n")
```

23. Un profesor afirma que el promedio de su clase en un examen nacional es diferente al promedio
nacional de 75 puntos. Una muestra de 36 estudiantes de su clase tiene una media de 78. La
desviación estándar poblacional es σ = 12. Con un nivel de significancia: α = 0,05.

*   Plantee las hipótesis nula y alternativa.
*   Calcule el estadístico Z.
*   Determine la región crítica y la región de rechazo.
*   Concluya si se rechaza H0.


```python
# ==============================================================================
# EJERCICIO 23: Prueba de Hipótesis Z para la Media (Examen)
# Bilateral (μ ≠ 75)
# ==============================================================================

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# --- DATOS DEL PROBLEMA ---
MU_0 = 75.0     # Valor hipotetizado (puntos)
X_BAR = 78.0    # Media muestral (puntos)
SIGMA = 12.0    # Desviación estándar poblacional (puntos)
N = 36          # Tamaño de la muestra
ALPHA = 0.05    # Nivel de significancia (5%)
ALPHA_MEDIOS = ALPHA / 2

print("--- EJERCICIO 23: Prueba Z (Promedio de Clase) ---")

# 1. HIPÓTESIS
print("\n1. Hipótesis:")
print(f"   H0 (Nula): μ = {MU_0} puntos")
print(f"   H1 (Alternativa): μ ≠ {MU_0} puntos (Bilateral)")

# 2. ESTADÍSTICO Z
SE = SIGMA / np.sqrt(N)
Z_ESTADISTICO = (X_BAR - MU_0) / SE
print(f"\n2. Estadístico Z Calculado:")
print(f"   Error Estándar (SE): {SE:.4f} puntos")
print(f"   Z Calculado: {Z_ESTADISTICO:.4f}")

# 3. REGIÓN CRÍTICA
# Z crítico para bilateral (colas α/2=0.025)
Z_CRITICO_POSITIVO = stats.norm.ppf(1 - ALPHA_MEDIOS)
Z_CRITICO_NEGATIVO = -Z_CRITICO_POSITIVO
print(f"\n3. Región Crítica (α = {ALPHA}):")
print(f"   Valores Críticos Z: $\\pm{Z_CRITICO_POSITIVO:.4f}")
print(f"   Región de Rechazo (RR): Z < {Z_CRITICO_NEGATIVO:.4f} ó Z > {Z_CRITICO_POSITIVO:.4f}")

# 4. CONCLUSIÓN
RECHAZA_H0 = (Z_ESTADISTICO < Z_CRITICO_NEGATIVO) or (Z_ESTADISTICO > Z_CRITICO_POSITIVO)
DECISION = "RECHAZAR H0" if RECHAZA_H0 else "NO RECHAZAR H0"
CONCLUSION = "Existe evidencia suficiente para concluir que el promedio de la clase es diferente a 75 puntos." if RECHAZA_H0 else "No existe evidencia suficiente para concluir que el promedio de la clase es diferente a 75 puntos."

print(f"\n4. Decisión: {DECISION}")
print(f"   {CONCLUSION}")

# GRÁFICA DE LA PRUEBA DE HIPÓTESIS
rango = np.linspace(-4, 4, 1000)
pdf = stats.norm.pdf(rango)

plt.figure(figsize=(10, 6))
plt.plot(rango, pdf, color='black', label='Distribución Normal Estándar (Z)')

# Región de Rechazo (colas)
rango_rechazo_neg = rango[rango < Z_CRITICO_NEGATIVO]
rango_rechazo_pos = rango[rango > Z_CRITICO_POSITIVO]
plt.fill_between(rango_rechazo_neg, stats.norm.pdf(rango_rechazo_neg), color='red', alpha=0.3, label=f'Región de Rechazo (α/2={ALPHA_MEDIOS})')
plt.fill_between(rango_rechazo_pos, stats.norm.pdf(rango_rechazo_pos), color='red', alpha=0.3)

# Marcar valores
plt.axvline(Z_CRITICO_NEGATIVO, color='red', linestyle='--', linewidth=2, label=f'Z Crítico Negativo = {Z_CRITICO_NEGATIVO:.4f}')
plt.axvline(Z_CRITICO_POSITIVO, color='red', linestyle='--', linewidth=2, label=f'Z Crítico Positivo = {Z_CRITICO_POSITIVO:.4f}')
plt.axvline(Z_ESTADISTICO, color='green', linestyle='-', linewidth=3, label=f'Z Calculado = {Z_ESTADISTICO:.4f}')

plt.title('Prueba de Hipótesis Z Bilateral (Promedio de Examen)')
plt.xlabel('Estadístico Z')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

print("----------------------------------------------------------\n")
```

26. Un restaurante afirma que su tiempo promedio de entrega a domicilio es 25 minutos. Un cliente sospecha que es mayor. Se registran 15 entregas, obteniendo una media de 28 minutos y desviación estándar muestral s = 4 minutos. Con un nivel de significancia de α = 0,05.

*   Plantee las hipótesis nula y alternativa.
*   Calcule el estadístico Z.
*   Determine la región crítica y la región de rechazo.
*   Concluya si se rechaza H0.

```python
# ==============================================================================
# EJERCICIO 26: Prueba de Hipótesis t para la Media (Tiempo de Entrega)
# Unilateral Derecha (μ > 25)
# n < 30 (n=15) y sigma desconocida (usamos s)
# ==============================================================================

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# --- DATOS DEL PROBLEMA ---
MU_0 = 25.0     # Valor hipotetizado (minutos)
X_BAR = 28.0    # Media muestral (minutos)
S = 4.0         # Desviación estándar muestral (minutos)
N = 15          # Tamaño de la muestra
ALPHA = 0.05    # Nivel de significancia (5%)
DF = N - 1      # Grados de libertad

print("--- EJERCICIO 26: Prueba t (Tiempo de Entrega) ---")

# 1. HIPÓTESIS
print("\n1. Hipótesis:")
print(f"   H0 (Nula): μ ≤ {MU_0} minutos")
print(f"   H1 (Alternativa): μ > {MU_0} minutos (Unilateral Derecha)")

# 2. ESTADÍSTICO t
SE = S / np.sqrt(N)
T_ESTADISTICO = (X_BAR - MU_0) / SE
print(f"\n2. Estadístico t Calculado:")
print(f"   Error Estándar (SE): {SE:.4f} minutos")
print(f"   t Calculado: {T_ESTADISTICO:.4f}")

# 3. REGIÓN CRÍTICA
# t crítico para unilateral derecha (1 - α)
T_CRITICO = stats.t.ppf(1 - ALPHA, DF)
print(f"\n3. Región Crítica (α = {ALPHA}):")
print(f"   Grados de Libertad (df): {DF}")
print(f"   Valor Crítico t: {T_CRITICO:.4f}")
print(f"   Región de Rechazo (RR): t > {T_CRITICO:.4f}")

# 4. CONCLUSIÓN
DECISION = "RECHAZAR H0" if T_ESTADISTICO > T_CRITICO else "NO RECHAZAR H0"
CONCLUSION = "Existe evidencia suficiente para concluir que el tiempo promedio de entrega a domicilio es mayor a 25 minutos." if DECISION == "RECHAZAR H0" else "No existe evidencia suficiente para concluir que el tiempo promedio de entrega a domicilio es mayor a 25 minutos."

print(f"\n4. Decisión: {DECISION}")
print(f"   {CONCLUSION}")

# GRÁFICA DE LA PRUEBA DE HIPÓTESIS
rango = np.linspace(-4, 4, 1000)
pdf = stats.t.pdf(rango, df=DF)

plt.figure(figsize=(10, 6))
plt.plot(rango, pdf, color='black', label=f'Distribución t de Student (df={DF})')

# Región de Rechazo (cola derecha)
rango_rechazo = rango[rango > T_CRITICO]
plt.fill_between(rango_rechazo, stats.t.pdf(rango_rechazo, df=DF), color='red', alpha=0.3, label=f'Región de Rechazo (α={ALPHA})')

# Marcar valores
plt.axvline(T_CRITICO, color='red', linestyle='--', linewidth=2, label=f't Crítico = {T_CRITICO:.4f}')
plt.axvline(T_ESTADISTICO, color='green', linestyle='-', linewidth=3, label=f't Calculado = {T_ESTADISTICO:.4f}')

plt.title('Prueba de Hipótesis t Unilateral Derecha (Tiempo de Entrega)')
plt.xlabel('Estadístico t')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

print("----------------------------------------------------------\n")
```

31. Una compañía de telefonía afirma que la varianza del tiempo de duración de las llamadas es de σ2= 2,5 minutos2. Se toma una muestra de n = 16 llamadas y se obtiene una varianza muestral de s2= 4. ¿Hay evidencia al 5% de significancia de que la varianza poblacional difiere de 2.5?

*   Plantee las hipótesis nula y alternativa.
*   Calcule el estadístico Z.
*   Determine la región crítica y la región de rechazo.
*   Concluya si se rechaza H0.

```python
# ==============================================================================
# EJERCICIO 31: Prueba de Hipótesis Chi-cuadrado (Varianza de Llamadas)
# Bilateral (σ² ≠ 2.5)
# ==============================================================================

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# --- DATOS DEL PROBLEMA ---
SIGMA2_0 = 2.5  # Varianza poblacional hipotetizada (minutos^2)
S2 = 4.0        # Varianza muestral (minutos^2)
N = 16          # Tamaño de la muestra
ALPHA = 0.05    # Nivel de significancia (5%)
DF = N - 1      # Grados de libertad
ALPHA_MEDIOS = ALPHA / 2

print("--- EJERCICIO 31: Prueba χ² (Varianza de Llamadas) ---")

# 1. HIPÓTESIS
print("\n1. Hipótesis:")
print(f"   H0 (Nula): σ² = {SIGMA2_0} minutos²")
print(f"   H1 (Alternativa): σ² ≠ {SIGMA2_0} minutos² (Bilateral)")

# 2. ESTADÍSTICO χ²
# Fórmula: χ² = (n - 1) * s² / σ²₀
CHI2_ESTADISTICO = (N - 1) * S2 / SIGMA2_0
print(f"\n2. Estadístico χ² Calculado:")
print(f"   χ² Calculado: {CHI2_ESTADISTICO:.4f}")

# 3. REGIÓN CRÍTICA
# Valores críticos para bilateral: χ²_1-α/2 y χ²_α/2
CHI2_CRITICO_IZQ = stats.chi2.ppf(ALPHA_MEDIOS, DF)
CHI2_CRITICO_DER = stats.chi2.ppf(1 - ALPHA_MEDIOS, DF)
print(f"\n3. Región Crítica (α = {ALPHA}):")
print(f"   Grados de Libertad (df): {DF}")
print(f"   Valores Críticos χ²: {CHI2_CRITICO_IZQ:.4f} (izquierda) y {CHI2_CRITICO_DER:.4f} (derecha)")
print(f"   Región de Rechazo (RR): χ² < {CHI2_CRITICO_IZQ:.4f} ó χ² > {CHI2_CRITICO_DER:.4f}")

# 4. CONCLUSIÓN
RECHAZA_H0 = (CHI2_ESTADISTICO < CHI2_CRITICO_IZQ) or (CHI2_ESTADISTICO > CHI2_CRITICO_DER)
DECISION = "RECHAZAR H0" if RECHAZA_H0 else "NO RECHAZAR H0"
CONCLUSION = "Existe evidencia suficiente para concluir que la varianza poblacional de la duración de las llamadas difiere de 2.5 minutos²." if RECHAZA_H0 else "No existe evidencia suficiente para concluir que la varianza poblacional de la duración de las llamadas difiere de 2.5 minutos²."

print(f"\n4. Decisión: {DECISION}")
print(f"   {CONCLUSION}")

# GRÁFICA DE LA PRUEBA DE HIPÓTESIS
rango = np.linspace(stats.chi2.ppf(0.001, DF), stats.chi2.ppf(0.999, DF), 1000)
pdf = stats.chi2.pdf(rango, DF)

plt.figure(figsize=(10, 6))
plt.plot(rango, pdf, color='black', label=f'Distribución Chi-cuadrado (df={DF})')

# Región de Rechazo (colas)
rango_rechazo_izq = rango[rango < CHI2_CRITICO_IZQ]
rango_rechazo_der = rango[rango > CHI2_CRITICO_DER]
plt.fill_between(rango_rechazo_izq, stats.chi2.pdf(rango_rechazo_izq, DF), color='red', alpha=0.3, label=f'Región de Rechazo (α/2={ALPHA_MEDIOS})')
plt.fill_between(rango_rechazo_der, stats.chi2.pdf(rango_rechazo_der, DF), color='red', alpha=0.3)

# Marcar valores
plt.axvline(CHI2_CRITICO_IZQ, color='red', linestyle='--', linewidth=2)
plt.axvline(CHI2_CRITICO_DER, color='red', linestyle='--', linewidth=2, label=f'χ² Críticos = {CHI2_CRITICO_IZQ:.2f}, {CHI2_CRITICO_DER:.2f}')
plt.axvline(CHI2_ESTADISTICO, color='green', linestyle='-', linewidth=3, label=f'χ² Calculado = {CHI2_ESTADISTICO:.4f}')

plt.title('Prueba de Hipótesis Chi-cuadrado Bilateral (Varianza de Llamadas)')
plt.xlabel('Estadístico χ²')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

print("----------------------------------------------------------\n")
```

```python
# ==============================================================================
# EJERCICIO 32: Prueba de Hipótesis Chi-cuadrado (Varianza de Llenado)
# Unilateral Derecha (σ² > 1.2)
# ==============================================================================

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# --- DATOS DEL PROBLEMA ---
SIGMA2_0 = 1.2  # Varianza poblacional hipotetizada (ml^2)
S2 = 2.0        # Varianza muestral (ml^2)
N = 20          # Tamaño de la muestra
ALPHA = 0.01    # Nivel de significancia (1%)
DF = N - 1      # Grados de libertad

print("--- EJERCICIO 32: Prueba χ² (Varianza de Llenado) ---")

# 1. HIPÓTESIS
print("\n1. Hipótesis:")
print(f"   H0 (Nula): σ² ≤ {SIGMA2_0} ml²")
print(f"   H1 (Alternativa): σ² > {SIGMA2_0} ml² (Unilateral Derecha)")

# 2. ESTADÍSTICO χ²
# Fórmula: χ² = (n - 1) * s² / σ²₀
CHI2_ESTADISTICO = (N - 1) * S2 / SIGMA2_0
print(f"\n2. Estadístico χ² Calculado:")
print(f"   χ² Calculado: {CHI2_ESTADISTICO:.4f}")

# 3. REGIÓN CRÍTICA
# Valor crítico para unilateral derecha: χ²_1-α
CHI2_CRITICO_DER = stats.chi2.ppf(1 - ALPHA, DF)
print(f"\n3. Región Crítica (α = {ALPHA}):")
print(f"   Grados de Libertad (df): {DF}")
print(f"   Valor Crítico χ²: {CHI2_CRITICO_DER:.4f}")
print(f"   Región de Rechazo (RR): χ² > {CHI2_CRITICO_DER:.4f}")

# 4. CONCLUSIÓN
DECISION = "RECHAZAR H0" if CHI2_ESTADISTICO > CHI2_CRITICO_DER else "NO RECHAZAR H0"
CONCLUSION = "Existe evidencia suficiente para concluir que la varianza poblacional del llenado es mayor a 1.2 ml²." if DECISION == "RECHAZAR H0" else "No existe evidencia suficiente para concluir que la varianza poblacional del llenado es mayor a 1.2 ml²."

print(f"\n4. Decisión: {DECISION}")
print(f"   {CONCLUSION}")

# GRÁFICA DE LA PRUEBA DE HIPÓTESIS
rango = np.linspace(stats.chi2.ppf(0.001, DF), stats.chi2.ppf(0.999, DF), 1000)
pdf = stats.chi2.pdf(rango, DF)

plt.figure(figsize=(10, 6))
plt.plot(rango, pdf, color='black', label=f'Distribución Chi-cuadrado (df={DF})')

# Región de Rechazo (cola derecha)
rango_rechazo_der = rango[rango > CHI2_CRITICO_DER]
plt.fill_between(rango_rechazo_der, stats.chi2.pdf(rango_rechazo_der, DF), color='red', alpha=0.3, label=f'Región de Rechazo (α={ALPHA})')

# Marcar valores
plt.axvline(CHI2_CRITICO_DER, color='red', linestyle='--', linewidth=2, label=f'χ² Crítico = {CHI2_CRITICO_DER:.4f}')
plt.axvline(CHI2_ESTADISTICO, color='green', linestyle='-', linewidth=3, label=f'χ² Calculado = {CHI2_ESTADISTICO:.4f}')

plt.title('Prueba de Hipótesis Chi-cuadrado Unilateral Derecha (Varianza de Llenado)')
plt.xlabel('Estadístico χ²')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

print("----------------------------------------------------------\n")
```

33. El control de calidad de una fábrica sostiene que la varianza en el grosor de piezas metálicas es de al menos σ2= 0,5 mm2. Una muestra de n = 12 piezas dio una varianza s2= 0,3. ¿Se puede concluir al 5% que la varianza poblacional es menor que la garantizada?

*   Plantee las hipótesis nula y alternativa.
*   Calcule el estadístico Z.
*   Determine la región crítica y la región de rechazo.
*   Concluya si se rechaza H0.

```python
# ==============================================================================
# EJERCICIO 33: Prueba de Hipótesis Chi-cuadrado (Varianza del Grosor)
# Unilateral Izquierda (σ² < 0.5)
# ==============================================================================

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# --- DATOS DEL PROBLEMA ---
SIGMA2_0 = 0.5  # Varianza poblacional hipotetizada (mm^2)
S2 = 0.3        # Varianza muestral (mm^2)
N = 12          # Tamaño de la muestra
ALPHA = 0.05    # Nivel de significancia (5%)
DF = N - 1      # Grados de libertad

print("--- EJERCICIO 33: Prueba χ² (Varianza del Grosor) ---")

# 1. PLANTÉE LAS HIPÓTESIS NULA Y ALTERNATIVA
print("\n1. Hipótesis:")
print(f"   H0 (Nula): σ² ≥ {SIGMA2_0} mm² (La varianza es al menos la garantizada)")
print(f"   H1 (Alternativa): σ² < {SIGMA2_0} mm² (La varianza es menor a la garantizada, Unilateral Izquierda)")

# 2. CALCULE EL ESTADÍSTICO χ²
# Fórmula: χ² = (n - 1) * s² / σ²₀
CHI2_ESTADISTICO = (N - 1) * S2 / SIGMA2_0
print(f"\n2. Estadístico χ² Calculado:")
print(f"   χ² Calculado: {CHI2_ESTADISTICO:.4f}")

# 3. DETERMINE LOS GRADOS DE LIBERTAD Y EL VALOR CRÍTICO, ADEMÁS DE LA REGIÓN CRÍTICA Y LA REGIÓN DE RECHAZO.
# Valor crítico para unilateral izquierda: χ²_α
CHI2_CRITICO_IZQ = stats.chi2.ppf(ALPHA, DF)
print(f"\n3. Región Crítica (α = {ALPHA}):")
print(f"   Grados de Libertad (df): {DF}")
print(f"   Valor Crítico χ²: {CHI2_CRITICO_IZQ:.4f}")
print(f"   Región de Rechazo (RR): χ² < {CHI2_CRITICO_IZQ:.4f}")

# 4. CONCLUYA SI SE RECHAZA H0
DECISION = "RECHAZAR H0" if CHI2_ESTADISTICO < CHI2_CRITICO_IZQ else "NO RECHAZAR H0"
CONCLUSION = "Existe evidencia suficiente para concluir que la varianza poblacional del grosor de las piezas es menor a la garantizada (0.5 mm²)." if DECISION == "RECHAZAR H0" else "No existe evidencia suficiente para concluir que la varianza poblacional del grosor de las piezas es menor a la garantizada (0.5 mm²)."

print(f"\n4. Decisión: {DECISION} (Dado que {CHI2_ESTADISTICO:.4f} es {'< (menor que)' if CHI2_ESTADISTICO < CHI2_CRITICO_IZQ else '≥ (mayor o igual a)'} {CHI2_CRITICO_IZQ:.4f})")
print(f"   {CONCLUSION}")

# GRÁFICA DE LA PRUEBA DE HIPÓTESIS
# Se ajusta el rango para visualizar mejor la cola izquierda
rango = np.linspace(stats.chi2.ppf(0.001, DF), stats.chi2.ppf(0.999, DF), 1000)
pdf = stats.chi2.pdf(rango, DF)

plt.figure(figsize=(10, 6))
plt.plot(rango, pdf, color='black', label=f'Distribución Chi-cuadrado (df={DF})')

# Región de Rechazo (cola izquierda)
rango_rechazo_izq = rango[rango < CHI2_CRITICO_IZQ]
plt.fill_between(rango_rechazo_izq, stats.chi2.pdf(rango_rechazo_izq, DF), color='red', alpha=0.3, label=f'Región de Rechazo (α={ALPHA})')

# Marcar valores
plt.axvline(CHI2_CRITICO_IZQ, color='red', linestyle='--', linewidth=2, label=f'χ² Crítico = {CHI2_CRITICO_IZQ:.4f}')
plt.axvline(CHI2_ESTADISTICO, color='green', linestyle='-', linewidth=3, label=f'χ² Calculado = {CHI2_ESTADISTICO:.4f}')

plt.title('Prueba de Hipótesis Chi-cuadrado Unilateral Izquierda (Varianza del Grosor)')
plt.xlabel('Estadístico χ²')
plt.ylabel('Densidad de Probabilidad')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

print("----------------------------------------------------------\n")
```
